# Mapping Sequences on Earth

```python
import geopandas as gpd
import pandas as pd
from lets_plot import *
import matplotlib.pyplot as plt
import numpy as np

LetsPlot.setup_html()
```

```python
from geopandas import GeoDataFrame


df_all = pd.read_csv('../../datasets/full_analysis/feat_all_seq.csv', usecols=['long_dec', 'lat_dec', 'group', 'host'])
# set long_dec and lat_dec as float
df_all['long_dec'] = df_all['long_dec'].astype(float)
df_all['lat_dec'] = df_all['lat_dec'].astype(float)
df_all['count'] = 1
df_all.reset_index(drop=True, inplace=True)
df_all.head()
```

```python
gdf_all = GeoDataFrame(df_all, geometry=gpd.points_from_xy(df_all.long_dec, df_all.lat_dec), crs='EPSG:4326')
zones = gdf_all.dissolve(by=['group'], aggfunc={'count': 'sum', 'host': 'first'})
zones.sort_values(by='count', ascending=False).head()
zones
```

```python
figure = ggplot() + geom_livemap() + \
  geom_point(aes(size='count', color='host'), data=zones, alpha=.5) + \
    ggtitle('Location of sequences included in the analysis') + \
    labs(size='Number of sequences', color='Host coral/sponge species') + \
    ggsize(1100, 800)
figure
```
